### Pertanyaan 2: Pada kondisi cuaca bagaimana pelanggan paling banyak menyewa sepeda?

```python
season_casual_users=day_df.groupby('season').casual_users.sum().sort_values(ascending=False).reset_index()
season_casual_users
```

```python
season_registered_users=day_df.groupby('season').registered_users.sum().sort_values(ascending=False).reset_index()
season_registered_users
```

```python
season_user=pd.merge(
    left=season_casual_users,
    right=season_registered_users,
    how='left',
    left_on='season',
    right_on='season'
)
season_user
```

```python
season_user_type=season_user.melt(id_vars='season',var_name='type users',value_name='count')
season_user_type
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x="season", y="count", hue="type users", data=season_user_type, palette="muted")
plt.ylabel(None)
plt.xlabel(None)
plt.title("Comparison of Number of Casual Users and Registered Users by Season")
casual_patch = mpatches.Patch(color=sns.color_palette("muted")[0], label='Casual User')
registered_patch = mpatches.Patch(color=sns.color_palette("muted")[1], label='Registered User')
plt.legend(handles=[casual_patch, registered_patch], title="User Type")
plt.show()
```

Pengguna yang sudah ter registrasi mendominasi sebagai pengguna Bike Sharing pada setiap musim, baik untuk musim Fall, Summer, Winter dan Springer. Dengan musim Fall merupakan musim yang paling diminati pengguna untuk menggunakan Bike Sharing.

### Pertanyaan 4 : Jumlah pelanggan dari casual dan registered, manakah yang paling banyak menyewa sepeda?

##RFM Analysis

```python
rfm_df = day_df.groupby(by="season", as_index=False).agg({
    "date": "max", # retrieve the date of the last order
    "instant": "nunique", # calculate the order quantity
    "total_users": "sum" # calculate the amount of revenue generated
})

rfm_df.columns = ["season", "max_order_timestamp", "frequency", "monetary"]

# calculates when the customer last made a transaction (days)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = day_df["date"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30,6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]
bar_width = 0.5
sns.barplot(y="recency", x="season", data=rfm_df.sort_values(by="recency",  ascending=True).head(5), palette=colors, hue="season", ax=ax[0], width=bar_width)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=20)

sns.barplot(y="frequency", x="season", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, hue="season", ax=ax[1], width=bar_width)
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=20)

sns.barplot(y="monetary", x="season", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, hue="season", ax=ax[2],  width=bar_width)
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=20)

plt.suptitle("Best Customer Based on RFM Parameters (season)", fontsize=20)
plt.show()
```

Analisis RFM menjelaskan mengenai Recency (Ketepatan), Frequency(Frekuensi) dan Monetary (Moneter). Berikut merupakan analisis RFM berdasarkan dataset Bike Sharing
*   Frequency (Frekuensi) menunjukkan jumlah transaksi yang terjadi pada ke 4 musim. Pada musim Fall terdapat 188 transaksi, Springer 181 transaksi, Summer 184 transaksi dan Winter 178.Dengan demikian, kita dapat melihat bahwa pengguna paling sering menggunakan Bike Sharing pada musim Fall   
*   Monetary (Moneter) menunjukkan jumlah total dari nilai transaksi Bike Sharing selama musim yang terjadi (Fall, Springer, Winter dan Summer). Pada musim Fall, total transaksi mencapai 1,061,129 sebagai transaksi paling banyak dari 4 musim tersebut.  
- Recency (Ketepatan) menunjukkan seberapa baru transaksi terakhir dilakukan selama musim tersebut. Pada musim gugur, transaksi terakhir memiliki 100 rekam jejak, musim semi tidak terdapat transaksi baru dan musimdingin memiliki 11 transaksi terbaru. Sedangkan pada musim Summer memiliki transaksi baru paling banyak yaitu sebanyak 194 rekam jejak.
